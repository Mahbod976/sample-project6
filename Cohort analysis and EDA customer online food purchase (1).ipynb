# Cohort analysis

```python
def get_month(x):
    return dt.datetime(x.year, x.month,1)
df3['hazf roz'] = df3['date'].apply(get_month)
df3
```

```python
df3['year'] = df3['date'].dt.year
df3['month'] = df3['date'].dt.month
df3['day'] = df3['date'].dt.day
```

```python
df3['Cohort Month'] =  df3.groupby('user_id')['hazf roz'].transform('min')
df3
```

```python
def get_date_elements(df, column):
    day = df[column].dt.day
    month = df[column].dt.month
    year = df[column].dt.year
    return day, month, year 
```

```python
_,Order_month,Order_year =  get_date_elements(df3,'date')
_,Cohort_month,Cohort_year =  get_date_elements(df3,'Cohort Month')
```

### plus <mark>1</mark> baraye inke mah 0 nadashte bashim

```python
year_diff = Order_year - Cohort_year
month_diff = Order_month - Cohort_month
df3['Cohort index'] = year_diff*12+month_diff+1
df3.head(100)
```

```python
cohort_data = df3.groupby(['Cohort Month', 'Cohort index'])['user_id'].apply(pd.Series.nunique).reset_index()
cohort_data
```

```python
cohort_table = cohort_data.pivot(index='Cohort Month', columns=['Cohort index'],values='user_id')
cohort_table
```
