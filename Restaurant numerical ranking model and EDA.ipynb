# KPI : customer satisfaction score
## 1.Average Customer Satisfaction Score
## 2.Weighted Customer Satisfaction Score
## 3. Normalized Customer Satisfaction Score

در سه حالت شاخص `نرمالایز شده `و‍‍`شاخص میانگین شده` و `میانگین وزن دار شده ‍`شاخص رضایت مشتری را در کل رستوران های شهر با پلتفرم اسنپ فود و با پلتفرم زودکس مقایسه میکنیم

حداقل و حداکثر نمره در پلتفرم ها یکسان و بین ۰ تا ۵ است و مقایسه میانگین معنادار میباشد

بدلیل بالاتر بودن نمره رستورانهایی که تعداد بالاتری نظر داشتند در دیتای گوگل مپ در این حالت با نرمالایز کردن آن، شاخص کل شهر بالاتر از اسنپ فود قرار گرفته

در حالت محاسبه میانگین بدون توجه به تعداد اسنپ فود نمره بالاتری کسب کرده که چون وزن تعداد نظرات در نظر گرفته نشده در حالت نرمالایز روایی بیشتری برای نتیجه میتوان قائل شد.

در حالت میانگین وزن دار، اهمیت بیشتری به کیفیت نطرات نسبت به تعداد نظرات قائل میشویم



```python
def calculate_normalized_score(df):
    weighted_rate = df['restaurant_rate'] * (df['restaurant_rate_count'] / df['restaurant_rate_count'].sum())
    normalized_score = weighted_rate.mean()
    return normalized_score
```

```python
score_google = calculate_normalized_score(df_google)
score_snappfood = calculate_normalized_score(df)
score_zoodex = calculate_normalized_score(df_zoodex)

print(f"Normalized Customer Satisfaction Score for all rafsanjan restaurants: {score_google:.4f}")
print(f"Normalized Customer Satisfaction Score for snappfood: {score_snappfood:.4f}")
print(f"Normalized Customer Satisfaction Score for zoodex: {score_zoodex:.4f}")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(['all rafsanjan restaurants', 'snappfood', 'zoodex'], [score_google, score_snappfood, score_zoodex], color=['blue', 'green', 'orange'])
plt.title('Normalized Customer Satisfaction Scores')
plt.xlabel('DataFrames')
plt.ylabel('Normalized Satisfaction Score')
plt.ylim(0, 0.5)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
avg_score_google = df_google['restaurant_rate'].mean()
avg_score_snappfood = df['restaurant_rate'].mean()
avg_score_zoodex = df_zoodex['restaurant_rate'].mean()

# Print scores
print(f"Average Customer Satisfaction Score for all rafsanjan restaurants: {avg_score_google:.2f}")
print(f"Average Customer Satisfaction Score for snappfood: {avg_score_snappfood:.2f}")
print(f"Average Customer Satisfaction Score for zoodex: {avg_score_zoodex:.2f}")
```

```python
weight_rate = 0.7
weight_rate_count = 0.3

def calculate_weighted_score(df):
    weighted_score = (df['restaurant_rate'] * weight_rate) + (df['restaurant_rate_count'] * weight_rate_count)
    return weighted_score.mean()

score_google = calculate_weighted_score(df_google)
score_snappfood = calculate_weighted_score(df)
score_zoodex = calculate_weighted_score(df_zoodex)


print(f"Weighted Customer Satisfaction Score for all rafsanjan restaurants: {score_google:.2f}")
print(f"Weighted Customer Satisfaction Score for snappfood: {score_snappfood:.2f}")
print(f"Weighted Customer Satisfaction Score for zoodex: {score_zoodex:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(['all rafsanjan restaurants', 'snappfood', 'zoodex'], [score_google, score_snappfood, score_zoodex], color=['blue', 'green', 'orange'])
plt.title('Weighted Customer Satisfaction Scores')
plt.xlabel('DataFrames')
plt.ylabel('Weighted Satisfaction Score')
plt.ylim(0, 70)  # Assuming scores are between 0 and 5
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# label resaurants in each dataset from a to e 


در جذب رستوران های جدید به سامانه اسنپ فود باید به رستورانهای با لیبل بالاتر در دیتاست کل رستوران های شهر توجه کرد و با دادن مشوق های گوناگون رستوران های برتر را جذب پلتفرم کرد

```python
weight_rate = 0.7
weight_rate_count = 0.3


def label_restaurants(df):
   
    grouped = df.groupby('restaurant_name').agg({
        'restaurant_rate': lambda x: (weight_rate * x.mean()),
        'restaurant_rate_count': lambda x: (weight_rate_count * x.mean())
    }).reset_index()

    grouped['weighted_score'] = grouped['restaurant_rate'] + grouped['restaurant_rate_count']

    grouped_sorted = grouped.sort_values(by='weighted_score', ascending=False).reset_index(drop=True)

    labels = ['A', 'B', 'C', 'D', 'E']
    grouped_sorted['label'] = pd.Categorical(pd.cut(grouped_sorted.index, bins=len(labels), labels=labels, include_lowest=True))
    
    return grouped_sorted
df_zoodex_labeled = label_restaurants(df_zoodex.copy()) 
df_google_labeled = label_restaurants(df_google.copy())
df_main_labeled = label_restaurants(df.copy())
```
