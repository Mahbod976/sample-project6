```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_org = pd.read_excel('Book2.xlsx')
data_org
```

### اضافه کردن ستون هفته


```python
data = data_org.copy()
```

```python
data.Date = pd.to_datetime(data.Date)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.insert(0,'week_of_year', data.Date.dt.isocalendar().week)
data
```

### محاسبه مقادیر شاخصه ها

```python
data['tamas_az_dast_rafte'] = data.Abandoned/data.Calls_Received
data['tamas_pasokh_dade_shode'] = data.Calls_Handled/data.Calls_Received
data['SL60'] = data.Handled_call_under_60s/data.Calls_Received
data['SL90'] = data.Handled_call_under_90s/data.Calls_Handled
data['FCR'] = data.Calls_Handled_by_agent/data.Calls_Handled
data['CSAT'] = data.Satisfactory_rate/(data.Dissatisfactory_rate + data.Normal_rate + data.Satisfactory_rate)
data
```

```python
data_grp = data.groupby('Pilot')
```

```python
data_grp.describe()
```

```python
fig, axs = plt.subplots(6, figsize=(20, 30), sharex=False)
for i, feature in enumerate(data.columns[20:]):
    for pilot, group_data in data_grp:
        axs[i].plot(group_data['Date'], group_data[feature], label=pilot)
    axs[i].set_title(feature)
    axs[i].legend()


fig.text(0.5, 0.07, 'Date', ha='center', fontsize=14);
```

```python
data_grp2 = data.groupby(['Pilot', 'week_of_year']).mean()
data_grp2
```

```python
fig, axs = plt.subplots(6, figsize=(20, 30), sharex=False)
for i, feature in enumerate(data.columns[20:]):
    for pilot in data['Pilot'].unique():
        pilot_data = data_grp2.loc[pilot]
        axs[i].plot(pilot_data.index.get_level_values('week_of_year'), pilot_data[feature], label=pilot)
    axs[i].set_title(feature)
    axs[i].legend()




# plt.tight_layout()
fig.text(0.5, 0.1, 'week_of_year', ha='center', fontsize=14);
```
